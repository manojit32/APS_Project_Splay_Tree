```python
import matplotlib.pyplot as plt
import sys
sys.setrecursionlimit(30002)
mmax=10000
bt=[]
st=[]
maxl=[]
while(mmax<26001):
    b=bst.BST()
    start=time.process_time()
    for i in range(1,mmax):
        b.insert(i)
    #for i in range(1,mmax):
    #    b.search(i)
    for i in range(mmax,i):
        b.remove(i)
    end=time.process_time()
    t1=end-start
    bt.append(t1)
    print(bt)
    s=splay.Splay()
    start=time.process_time()
    for i in range(1,mmax):
        s.insert(i)
    #for i in range(1,mmax):
     #   s.search(i)
    for i in range(mmax,1):
        s.remove(i)
    end=time.process_time()
    t1=end-start
    st.append(t1)
    print(st)
    maxl.append(mmax)
    mmax=mmax+2000    
plt.plot(maxl,bt,'o-')
plt.plot(maxl,st,'x-')
plt.show()
```
